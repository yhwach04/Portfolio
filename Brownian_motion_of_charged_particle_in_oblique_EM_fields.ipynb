# **Diffusion coefficients**
The effective one-demensional diffusion coefficient is given by
$$
D = \lim_{\tau\to∞} \frac{\sigma^2}{2\tau}
$$





## **Diffusion along the $x-$ axis**


Now for the diffusion coefficient along the $x-$ axis yields to


$$
D_x = \lim_{\tau\to∞} \frac{\sigma_x^2}{2\tau} = λk_B T\left[\frac{λ/m_y}{\left(\frac{qB}{\sqrt{m_x m_y}}\right)^2+\left(\frac{\lambda}{m_x m_y}\right)^2}\right]
$$

now we let $\lambda = 1$ and $qB = 0.2$, and normalize $k_B T$,

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

#Generates values for my
my_values = np.linspace(0.1, 2, 100)

#Define your function
def diffusion_x(mx, my):
  return (1/my)/((0.04/mx*my)+(1/mx*my)**2)

#Generates values for mx
#mx_vals0 = np.array([diffusion_x(0.0, my) for my in my_values])
mx_vals1 = np.array([diffusion_x(0.1, my) for my in my_values])
mx_vals2 = np.array([diffusion_x(0.2, my) for my in my_values])
mx_vals3 = np.array([diffusion_x(0.3, my) for my in my_values])
mx_vals4 = np.array([diffusion_x(0.4, my) for my in my_values])
mx_vals5 = np.array([diffusion_x(0.5, my) for my in my_values])


#colors
colors = sns.color_palette("magma", 5)
#seshadri = ['#c3121e', '#0348a1', '#ffb01c', '#027608', '0193b0']

#Plots the diffusion coefficient
fig, ax = plt.subplots(figsize=(5,5))
#ax.plot(my_vals, mx_vals0, 'g', label='m_x = 0.0')
ax.plot(my_values, mx_vals1, color=colors[0], label='$m_x = 0.1$')
ax.plot(my_values, mx_vals2, color=colors[1], label='$m_x = 0.2$')
ax.plot(my_values, mx_vals3, color=colors[2], label='$m_x = 0.3$')
ax.plot(my_values, mx_vals4, color=colors[3], label='$m_x = 0.4$')
ax.plot(my_values, mx_vals5, color=colors[4], label='$m_x = 0.5$')
plt.xlim(0, 2.0)
plt.ylim(0, 3.0)
plt.xlabel("$m_y$", fontsize=13)
plt.ylabel('$D_x$', fontsize=14)
plt.minorticks_on()
plt.tick_params(direction='in',right=True,top=True)
plt.tick_params(labelsize='14')
plt.tick_params(labelbottom=True, labeltop=False, labelright=False, labelleft=True)
plt.tick_params(direction='in', bottom=True)
# ax.ticklabel_format(axis='x', scilimits=(-0,1), useMathText=True)
# ax.ticklabel_format(axis='y', scilimits=(-0,1), useMathText=True)
#plt.xlim(0, -0.01)
plt.legend(fontsize=14)
plt.savefig('Dx.png', dpi=300, bbox_inches='tight')
plt.show()
```
