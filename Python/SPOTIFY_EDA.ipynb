# Convert the duration from milliseconds to seconds

```python
df['duration'] = df['duration_ms'].apply(lambda x: round(x/1000))
df.drop('duration_ms', inplace=True, axis=1)
```

```python
df.duration.head()
```

# Data Visulization

## Data Correlation

```python
corr = df.drop(['genre','artist_name', 'key','mode','track_name','track_id', 'time_signature'], axis=1).corr()
plt.figure(figsize=(14,6))
sns.heatmap(corr, annot=True, cmap='viridis', fmt='.1g', vmax=1, vmin=-1, center=0)
plt.title('Correlation Matrix')
plt.show()
```

## Scatter Plot Between Loudness and Energy

```python
sample = df.sample(int(0.005*len(df)))
plt.figure(figsize=(12,6))
sns.scatterplot(x='loudness', y='energy', data=sample)
plt.title('Loudness vs Energy')
plt.xlabel('Loudness')
plt.ylabel('Energy')
plt.show()
```

```python
sample = df.sample(int(0.005*len(df)))
plt.figure(figsize=(12,6))
sns.regplot(x='acousticness', y='popularity', data=sample)
plt.title('Popularity vs Acousticness')
plt.xlabel('Acousticness')
plt.ylabel('Popularity')
plt.show()
```

```python
sample = df.sample(int(0.005*len(df)))
plt.figure(figsize=(12,6))
sns.regplot(x='danceability', y='popularity', data=sample)
plt.title('Popularity vs Danceability')
plt.xlabel('Acousticness')
plt.ylabel('Popularity')
plt.show()
```
